```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
weather = pd.read_csv('Test.csv')
```

# BARPLOT

```python
sns.barplot(x='humidity', y='temperature', data=weather)
```

# HISTPLOT


```python
sns.histplot(weather['humidity'])
```

# HISTPLOT WITH KDE

```python
sns.histplot(weather['temperature'],kde=True);
```

# JOINTPLOT

```python
sns.jointplot(x='humidity',y='temperature',data=weather)
```

```python
sns.jointplot(x='humidity',y='temperature',data=weather, kind="hex")
```

```python
sns.jointplot(x=weather['humidity'],y=weather['temperature'], kind="kde")
```

# PAIRPLOT

```python
sns.pairplot(weather[['temperature','humidity','dew_point']])
```

# STRIPPLOT

```python
sns.stripplot(x=weather['weather_type'],y=weather['temperature'])
```

```python
sns.stripplot(x=weather['weather_type'],y=weather['temperature'],jitter=True)
```

# SWARMPLOT

```python
sns.swarmplot(x=weather['dew_point'],y=weather['temperature'],hue=weather['weather_type'],palette='rainbow')
```

# BOXPLOT

```python
sns.boxplot(x=weather['humidity'],y=weather['temperature'],hue=weather['weather_type'])
```

# BARPLOT

```python
sns.barplot(x=weather['humidity'],y=weather['temperature'],hue=weather['weather_type'])
```

# COUNTPLOT

```python
sns.countplot(x=weather['weather_type'])
```

# POINTPLOT

```python
sns.pointplot(x=weather['humidity'],y=weather['temperature'],hue=weather['weather_type'])
```

# VOILINPLOT

```python
sns.violinplot(x='wind_speed',y='temperature',data=weather,palette='rainbow')
```
